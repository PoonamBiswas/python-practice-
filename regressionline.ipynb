```python
import matplotlib.pyplot as plt
```

```python
#data
sizes = [1000, 1500, 2000, 2500, 3000]
prices = [300000, 450000, 600000, 750000, 900000]
```

```python
#graph
plt.Figure(figsize = (8,6))
plt.plot(sizes,prices,marker='o', linestyle = '-', color= 'orange', label = 'Price Trend')
#labels 
plt.xlabel('sizes')
plt.ylabel('price')
plt.title('Price Trend')
plt.legend()
plt.grid(True)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Reshaping the data into the right shape
sizes = np.array(sizes).reshape(-1, 1)  # Assuming sizes is a list or array
prices = np.array(prices)

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(sizes, prices)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Predict the prices using the regression line
predicted_prices = model.predict(sizes)

# Plot the regression line on top of the scatter plot
plt.plot(sizes, predicted_prices, color='blue', label='Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
r_squared = model.score(sizes, prices)
print(f"R-squared: {r_squared}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(prices, predicted_prices)
mse = mean_squared_error(prices, predicted_prices)
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
```
